```python
class User:
    def __init__(self, username, password, balance):
        self.username = username
        self.password = password
        self.balance = balance
    
    def check_balance(self):
        print(f"{self.username}, your current balance is {self.balance}")

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"{amount} is deposited successfully")
            self.check_balance()
        else:
            print("Deposit amount must be greater than 0")

    def withdraw(self, amount):
        if amount > self.balance:
            print("Not sufficient balance in your account")
            self.check_balance()
        else:
            self.balance -= amount
            print(f"{amount} is withdrawn successfully")
            self.check_balance()


class ATM:
    def __init__(self):
        self.users = {
            "Reaper" : User('Reaper','12345',5000),
            "Demigod" : User('Demigod','54321',2000)
        }

    def add_user(self, username, password, balance):
        if username in self.users:
            print(f"{username}, user already exists")
        else:
            self.users[username] = User(username, password, balance)
            print(f"{username}, user added successfully")

    def validate(self, username, password):
        if username in self.users and password == self.users[username].password:
            return self.users[username]
        else:
            print("Authentication failed, invalid credentials or user doesn't exist")
            return None

    def register_user(self):
        username = input("Enter username here: ")
        if username in self.users:
            print(f"{username}, user already exists")
        else:
            password = input("Enter your password here: ")
            initial_deposit = int(input("Enter the amount of your initial deposit: "))
            self.add_user(username, password, initial_deposit)


def main():
    atm = ATM()

    while True:
        print("\nWelcome to the ATM Machine\n1. Register\n2. Login\n3. Exit")
        option = input("Enter your choice: ")

        if option == "1":
            atm.register_user()
        elif option == "2":
            username = input("Enter your username: ")
            password = input("Enter your password: ")

            user = atm.validate(username, password)
            if user:
                while True:
                    print("\n1. Check Balance\n2. Deposit\n3. Withdraw\n4. Logout")
                    choice = input("Enter your choice: ")

                    if choice == "1":
                        user.check_balance()
                    elif choice == "2":
                        amount = int(input("Enter the amount to deposit: "))
                        user.deposit(amount)
                    elif choice == "3":
                        amount = int(input("Enter the amount to withdraw: "))
                        user.withdraw(amount)
                    elif choice == "4":
                        print("Logged out successfully")
                        break
                    else:
                        print("Invalid choice. Please try again.")
        elif option == "3":
            print("Thank you for using the ATM")
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
